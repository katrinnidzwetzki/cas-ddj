### Und nun hängen wir alls zusammen

```python
#eine leere df bauen
ff = pd.DataFrame()
headers = ['Gemeinde', 'Sachschaden', 'Personenschaden', 'Getötete', 
           'Schwerverletzte', 'Lebensbedrohlich Verletzte', 
           'Erheblich Verletzte', 'Leichtverletzte']

#Und nun durch alles iterieren
#https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html
for number in range(0,8):
    temp = tables_stream[number].df #
    temp = temp[9:]
    temp.columns = headers
    frames = [ff, temp]
    ff = pd.concat(frames)
```

```python
ff.info()
```

```python
## Mit leeren Strings umgehen
def clean(elem):
    if elem == "":
        return 0
    else:
        return elem
ff = ff.applymap(clean)
```

```python
## Die Gemeinden zwischenzeitlich löschen und dann wieder reinnehmen bei der Umwandlung in Zahlen
gem = ff['Gemeinde'] #variable, "kopie" von gemeinden
del ff['Gemeinde']   #vorübergehend loeschen
ff = ff.astype(int)
ff['Gemeinde'] = gem   ###neue spalte gemeinden
ff = ff[['Gemeinde', 'Sachschaden', 'Personenschaden', 'Getötete', 
           'Schwerverletzte', 'Lebensbedrohlich Verletzte', 
           'Erheblich Verletzte', 'Leichtverletzte']].copy() ##Reihenfolge, gemeinde an erste stelle
```

```python
#index anpassen
ff = ff.reset_index()   ### neuer index beginnend mit 1
del ff['index']  #loescht alte index-spalte
```
