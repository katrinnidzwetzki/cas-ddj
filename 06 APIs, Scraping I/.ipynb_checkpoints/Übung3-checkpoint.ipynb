#### 1. Importiere requests, BeautifulSoup und Pandas

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

#### 2. Lade den Inhalt des Parkhaus-RSS-Feeds in eine Variabel

```python
r = requests.get('http://www.plszh.ch/plsFeed/rss')
```

#### 3. Ziehen wir alle Daten heraus und machen daraus eine Liste

```python
daten = r.text
soup = BeautifulSoup(daten, 'xml')
namen = soup.find_all('title')
adressen = soup.find_all('link')
```

```python
adressen
```

```python
liste = []
for name, adresse in zip(namen,adressen):   #bringen zwei oder mehr objekte zusammen, gleiche anzahl zeilen!
    mini_dict = {'Parkhaus': name.text, 'Ort': adresse.text}   #  .text ohne html-befehle
    liste.append(mini_dict)


liste

```

```python
pd.DataFrame(liste)[1:]
```

#### 4. Ziehen wir alle Links heraus und machen daraus eine Liste

#### 5. Nun bauen wir eine Dictionary aus den beiden Listen mit dem ersten Key "Titel" und zweiten Key "URL".

#### 6. Schliesslich machen wir daraus einen DataFrame (eine für uns Menschen lesbare Tabelle) und speichern das Ergebnis ab.

```python
####### zweiter vesuch mit freien Parkplaetzen

daten = r.text
soup = BeautifulSoup(daten, 'xml')
namen = soup.find_all('title')
adressen = soup.find_all('link')
freie_plaetze = soup.find_all('description')

ergebnis = []
for name, adresse, freie_plaetze in zip(namen,adressen, freie_plaetze):   
    mini_dict = {'Parkhaus': name.text, 'Ort': adresse.text, 'Frei: ': freie_plaetze.text }  
    ergebnis.append(mini_dict)

ergebnis

```

```python
pd.DataFrame(ergebnis)[1:]    # [1:] für liste ohne erste zeile der ausgangsdaten
```
