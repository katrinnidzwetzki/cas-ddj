```python
tokens = word_tokenize(testtext)
tokens = [word.lower() for word in tokens if word.isalpha()]
testtext_clean = (" ").join(tokens)
#tokens
testtext_clean  #text ohne satzzeichen in kleinbuchstaben
```

```python
%matplotlib inline
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import os
from os import path
d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()

maske = np.array(Image.open(path.join(d, "maske.png")))

wc = WordCloud(background_color="white", max_words=2000, mask=maske, contour_width=1, contour_color='black',collocations=False, normalize_plurals=False)
wc.generate(textfuerbild)
plt.figure(figsize=(15,8))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
plt.savefig("test.png")
```
